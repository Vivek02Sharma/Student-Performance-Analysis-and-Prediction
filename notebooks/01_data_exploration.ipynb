#### 2. Data Distribution and Visualization

```python
# checking distribution using histogram 
plt.figure(figsize=(15,25))
for i,col in enumerate(df.drop(['StudentId','Remark','Grade','TotalMarksObtained'],axis = 1).columns,1):
        plt.subplot(10,2,i)
        sns.histplot(df[col],bins = 30,kde = True)
        plt.title(f"Histogram of {col}")
        
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
# checking the outliers
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
sns.boxplot(data =df, x = df['TotalMarksObtained'])
plt.title("Boxplot of TotalMarksObtained")
plt.subplot(1,2,2)
sns.boxplot(data =df, x = df['Percentage'])
plt.title("Boxplot of Percentage")
plt.show()
```

```python
# checking count of Remark and Grade columns using countplot
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.title("Count bar of Remark")
sns.countplot(x = df['Remark'],hue = df['Remark'])

plt.subplot(1,2,2)
plt.title("Count bar of Grade")
sns.countplot(x = df['Grade'],hue = df['Grade'])
plt.show()
```

```python
# checking pie chart of passing and failing
remark = df['Remark'].value_counts(normalize = True)

plt.pie(remark,labels = remark.index,autopct='%1.1f%%',startangle = 100)
plt.title('Percentage of Pass and Fail')
plt.axis('equal')
plt.show()
```

```python
# Precentage Distribution
sns.histplot(df['Percentage'],kde = True,bins = 30,color = 'salmon')
plt.title('Distribution of Percentage')
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.show()
```
