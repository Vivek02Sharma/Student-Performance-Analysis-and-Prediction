## 6. Outlier Detection

```python
from scipy import stats

df['TotalMarks_z'] = stats.zscore(df['TotalMarksObtained'])
df['Percentage_z'] = stats.zscore(df['Percentage'])
df['SGPA_z'] = stats.zscore(df['SGPA'])

# Define outliers as those with z-scores beyond Â±3
outliers_total_marks = df[df['TotalMarks_z'].abs() > 3]
outliers_percentage = df[df['Percentage_z'].abs() > 3]
outliers_sgpa = df[df['SGPA_z'].abs() > 3]

print('Outliers in Total Marks Obtained:')
print(outliers_total_marks)
print('Outliers in Percentage:')
print(outliers_percentage)
print('Outliers in SGPA:')
print(outliers_sgpa)
```
