```python
import pandas as pd
```

```python
df = pd.read_csv('BMS SEM II.csv')
df.head()
```

```python
# checking missing value 
df.isnull().sum()
```

```python
# checking duplicate value
df.duplicated().sum()
```

```python
# dropping the columns and rows where values are missing,duplicated and unnecessary
df.dropna(how = "all",axis = 1,inplace = True)
df.dropna(how = "all",inplace = True)
df.drop_duplicates(inplace = True)
df.drop(['CourseName','ExamRollNumber','StudentName'],inplace = True,axis = 1)
```

```python
df.head()
```

```python
# dividing columns based on data types
categorical_columns = []
numerical_columns = []
for column in df.columns:
        if df[column].dtypes == 'O':
            categorical_columns.append(column)
        else:
            numerical_columns.append(column)
```

```python
# this function add grace marks
def adding_grace(row):
    # if '*' present in row
    if '*' in row:
        num = row.split('*')
        return int(num[0]) + int(num[1])
    # '*' not present in row
    else:
        return int(row)
```

```python
# converting the data type of the column which contain grace('*') marks
def apply_grace_marks(categorical_columns):
    columns_with_grace = []
    for col in categorical_columns:
        if col.startswith('INT_') or col.startswith('EXT_'):
            columns_with_grace.append(col)
            
    for col in columns_with_grace:
        df[col] = df[col].apply(adding_grace)
    return df
```

```python
df = apply_grace_marks(categorical_columns)
df.head(20)
```

```python
# save the process dataset
df.to_csv('BMS_sem2_process.csv', index=False,header = True)
```
